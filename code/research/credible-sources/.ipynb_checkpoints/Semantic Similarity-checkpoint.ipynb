```python
import json
import retinasdk
```

```python
with open("./data/input.json", "r") as fp:
    claim_text = json.load(fp)["text"]
```

```python
with open("./results/er_opt.json", "r") as fp:
    df = json.load(fp)
    for i in range(len(df["articles"])):
        df["articles"][i] = json.loads(df["articles"][i])
```

```python
fullClient = retinasdk.FullClient("79a39910-4ff4-11e9-8f72-af685da1b20e", apiServer="http://api.cortical.io/rest", retinaName="en_associative")
```

```python
for article in df["articles"]:
    print(fullClient.compare(json.dumps([{"text": claim_text}, {"text": article["body"]}])))
```

```python
for article in df["articles"]:
    print(fullClient.compareImage(json.dumps([{"text": claim_text}, {"text": article["body"]}])))
```

```python
fh = open("./results/imageToSave.png", "wb")
fh.write(fullClient.compareImage(json.dumps([{"text": claim_text}, {"text": article["body"]}])).decode('base64'))
fh.close()
```
